```python
## CONCAT 

india_weather=pd.DataFrame({
    'city':['AP','telangana','karnataka'],
     'temp':[22,66,55],
      'humidity':[91,55,58]
    
})
india_weather
```

```python
us_weather=pd.DataFrame({
    'city':['newyork','AP','lonova'],
     'temp':[44,55,39],
      'humidity':[87,74,58]
    
})
us_weather
```

```python
df=pd.concat([india_weather,us_weather],ignore_index=True)
df

df=pd.concat([india_weather,us_weather],keys=['india','us'])
df
```

```python
df.loc['india']
```

```python
## merge is similar to joins 
df1=pd.merge(india_weather,us_weather,on='city')
df1
```

```python
## outer join har an argument how='left ' is for left join,right for right join
df2=pd.merge(india_weather,us_weather,on='city',how='right',indicator=True)
df2
```

```python
df = pd.DataFrame({'Name': {0: 'John', 1: 'Bob', 2: 'Shiela'},
                   'Course': {0: 'Masters', 1: 'Graduate', 2: 'Graduate'},
                   'Age': {0: 27, 1: 23, 2: 21}})
df
```

```python
df1=pd.melt(df,id_vars=["Name"],value_vars=['Age','Course'])
df1=pd.melt(df,id_vars=["Name"],value_vars=['Course'],var_name='Description',value_name='Course_col')
df1=pd.melt(df,id_vars=["Name"],value_vars=['Age','Course'],ignore_index=True)

df1
```

Stack and unstack



```python
df = pd.DataFrame({'Name': {0: 'John', 1: 'Bob', 2: 'Shiela'},
                   'Course': {0: 'Masters', 1: 'Graduate', 2: 'Graduate'},
                   'Age': {0: 27, 1: 23, 2: 21}})
df
```

```python
stacked_df=df.stack()
stacked_df=df.stack(level=0)
stacked_df
```

```python
unstacked_df=stacked_df.unstack()
unstacked_df
```
