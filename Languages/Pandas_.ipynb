# Mathematical operation on series

we can use operators or methods for mathematical computaions
it perform action on each value of one series w.r.t othe series.values
we will get negative also

```python
# 1)  add ( + )  ,  we can used both add methos and "+" operator for addition

sd1=pd.Series([10,20,30,40,50,60])
sd2=pd.Series([6,5,4,3,2,1])
print(sd1)
print(sd2)
print()
print("Addition of two series is : ",sd1.add(sd2),sd1+sd2)
```

```python
print("Subtraction of the series is: ",sd1.subtract(sd2),sd2-sd1)
```

```python
print("Multiplication of two series is ",sd1*sd2,sd1.multiply(sd2))
```

```python
print("Division of two series is :",sd1.divide(sd2),sd2/sd1)
```

```python
# print(sd1)
# print(sd2)
print("Floor division of two series is : ",sd1%sd2)   # it gives the reminder
print(sd1//sd2)      # it gives the quotient without decimal places
print(sd1/sd2)      # it gives quotient with decamal values
```

```python
print("Is sd1 greater than sd2 ",sd1>sd2,sd1.gt(sd2))
print()
print("Is sd1 lessthan sd2 :",sd1<sd2,sd1.le(sd2),sd1.lt(sd2))
print()
print("Is sd1 equal to sd2 :",sd1==sd2)
print()
print("The power is :", sd1**sd2)
```

# dropna   &  fillna

df.dropna(axis=0,how ="any")

axis : is used to delete based on column or row by default axis=0 means any row if one value is NaN then it dropd the entire row
      we can use 1  to delete based on column data

how : it means how the deletion procees need to happen weather entire row or column having NaN value of any one value NaN 
      by default it how is "any" 
      


```python
data=pd.read_excel("D:\\vscode\\pandas\\missing_data.xlsx")
df=pd.DataFrame(data)
df
```

```python
df.dropna(axis=1,how="all")    # this will doesn't effect in  original df ,for that we need to use " inplace=True " parameter 
```
